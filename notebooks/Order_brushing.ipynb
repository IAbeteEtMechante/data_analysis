```python
shop_to_user = defaultdict(set)
```

```python
count = 0
df.groupby('shopid', as_index=False).apply(get_suspicious_buyers)
```

```python
def get_suspicious_buyers(df_shop):
#     df_shop = df.groupby('shopid', as_index=False).get_group(10402)
    df_shop = df_shop.reset_index(drop = True)
    N = len(df_shop)
    df_shop['suspicious_buyer'] = 0
    shop = df_shop.loc[0,"shopid"]
    global count
    count += 1
    print(f"{count}/18700")
    for start_idx in range(N):
        start = df_shop.loc[start_idx,var.event_time]
        end = start + 3600
        df_hour = df_shop.loc[(df_shop[var.event_time] >= start) & (df_shop[var.event_time] <= end)]
        n_orders = df_hour.shape[0]
        n_unique_buyers = df_hour[var.userid].nunique()
        concentrate_rate = n_orders/n_unique_buyers
        if concentrate_rate >= 3:
    #         df_hour.loc[:,'suspicious_buyer'] = 1
    #         print(df_hour)
    #         print(df_shop.loc[4])
            df_shop.loc[(df_shop[var.event_time] >= start) & (df_shop[var.event_time] <= end), 'suspicious_buyer'] = 1
            break
    df_shop = df_shop[df_shop['suspicious_buyer'] == 1]
    df_shop['userid'].value_counts()
    occur_dict = df_shop['userid'].value_counts().to_dict()
    if occur_dict:
        MAX = max(occur_dict.values())
        for key in occur_dict.keys():
            if occur_dict[key] == MAX:
                shop_to_user[shop].add(key)
```
